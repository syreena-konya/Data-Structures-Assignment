Algorithm Analysis and Complexity

Big-O : Represents the upper bound of an algorithm's running time. Describes the worst case scenario.

Big-Theta : Represents the tight bound. Defines the complexity if the upper and lower bounds are the same.

Big-Omega : Represents the lower bound. Describes the best case scenario.

Time Analysis *O*(n*n) : Because there is a nested loop where each loop runs n times(n*n)

Efficiency: The algorithm is not efficient for large values of n. As n grows the execution time increases.

```python
arr = [21, 10, 45, 32, 67, 5, 18]

# a) Linear Search
def linear_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return i
    return -1

# b) Binary Search (Requires sorted list)
def binary_search(data, target):
    low = 0
    high = len(data) - 1
    while low <= high:
        mid = (low + high) // 2
        if data[mid] == target:
            return mid
        elif data[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# c) Bubble Sort
def bubble_sort(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

# Execution
print(f"Linear Search (32) Index: {linear_search(arr, 32)}")
sorted_arr = bubble_sort(arr.copy())
print(f"Sorted Array: {sorted_arr}")
print(f"Binary Search (45) Index: {binary_search(sorted_arr, 45)}")
```

```python
# a) Node Class
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# b) LinkedList Class
class LinkedList:
    def __init__(self):
        self.head = None

    def insert_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")
```

Applications of Data Structures

The most appropriate data structure is a queue.
A queue follows a first in-first out principle.
In a finance office setting,the first student to arrive should be the first on to be attended to.

```python
from collections import deque

class FinanceQueue:
    def __init__(self):
        self.queue = deque()

    def add_student(self, name):
        self.queue.append(name)
        print(f"Student {name} added to the queue.")

    def serve_student(self):
        if self.queue:
            removed = self.queue.popleft()
            print(f"Serving student: {removed}")
        else:
            print("No students in the queue.")

    def display_queue(self):
        print("Current Queue:", list(self.queue))

# Simulation
office = FinanceQueue()
office.add_student("Alice")
office.add_student("Bob")
office.display_queue()
office.serve_student()
office.display_queue()
```
